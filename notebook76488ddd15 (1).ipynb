```python
import cv2
import os
from tqdm import tqdm
import numpy as np


base_dir = "/kaggle/working/video_dataset/Shop DataSet"
output_dir = "/kaggle/working/frames_dataset"
os.makedirs(output_dir, exist_ok=True)

NUM_FRAMES = 16  

def extract_frames(video_path, out_folder, num_frames=NUM_FRAMES):
    os.makedirs(out_folder, exist_ok=True)
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if total_frames == 0:
        return
    frame_indices = np.linspace(0, total_frames - 1, num_frames, dtype=int)
    idx = 0
    for i in range(total_frames):
        ret, frame = cap.read()
        if not ret:
            break
        if i in frame_indices:
            frame = cv2.resize(frame, (224, 224))
            cv2.imwrite(os.path.join(out_folder, f"frame_{idx:03d}.jpg"), frame)
            idx += 1
    cap.release()

for label in ["shop lifters", "non shop lifters"]:
    videos_path = os.path.join(base_dir, label)
    for video_file in tqdm(os.listdir(videos_path), desc=f"Processing {label}"):
        video_path = os.path.join(videos_path, video_file)
        if not video_file.lower().endswith((".mp4", ".avi", ".mov")):
            continue
        out_folder = os.path.join(output_dir, label, os.path.splitext(video_file)[0])
        extract_frames(video_path, out_folder)

print("✅ Done extracting frames!")
```

```python
from transformers import AutoImageProcessor, TimesformerForVideoClassification
import torch

model_name = "facebook/timesformer-base-finetuned-k400"
processor = AutoImageProcessor.from_pretrained(model_name)
model = TimesformerForVideoClassification.from_pretrained(model_name)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

print("✅ Model Loaded Successfully:", model_name)
```

```python
import cv2
import numpy as np
from PIL import Image

def load_video_frames(video_path, num_frames=8):

    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_indices = np.linspace(0, total_frames - 1, num_frames, dtype=int)
    frames = []

    for idx in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        success, frame = cap.read()
        if not success:
            break
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frames.append(Image.fromarray(frame))
    cap.release()
    return frames
```

```python
import torch

video_path = "/kaggle/working/video_dataset/Shop DataSet/shop lifters/shop_lifter_125.mp4"

frames = load_video_frames(video_path)
inputs = processor(frames, return_tensors="pt").to(device)

with torch.no_grad():
    outputs = model(**inputs)
    logits = outputs.logits
    predicted_label = logits.argmax(-1).item()


label_names = model.config.id2label
print("🎬 Prediction:", label_names[predicted_label])
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import os
import random

class VideoFramesDataset(Dataset):
    def __init__(self, root_dir, transform=None, frames_per_video=8):
        self.root_dir = root_dir
        self.transform = transform
        self.frames_per_video = frames_per_video

        self.samples = []
        self.classes = sorted(os.listdir(root_dir))  # ['non shop lifters', 'shop lifters']
        self.class_to_idx = {cls: i for i, cls in enumerate(self.classes)}

        for cls in self.classes:
            class_dir = os.path.join(root_dir, cls)
            for video_folder in os.listdir(class_dir):
                folder_path = os.path.join(class_dir, video_folder)
                if os.path.isdir(folder_path):
                    frames = sorted(os.listdir(folder_path))
                    if len(frames) >= self.frames_per_video:
                        self.samples.append((folder_path, self.class_to_idx[cls]))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        folder_path, label = self.samples[idx]
        frame_files = sorted(os.listdir(folder_path))
        selected_frames = random.sample(frame_files, self.frames_per_video)

        frames = []
        for f in selected_frames:
            img = Image.open(os.path.join(folder_path, f)).convert("RGB")
            if self.transform:
                img = self.transform(img)
            frames.append(img)

        frames = torch.stack(frames)  # [T, C, H, W]
        return frames, label

# 🔹 التحويلات (Transformations)
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

# 🔹 إنشاء الداتا
dataset_path = "/kaggle/working/frames_dataset"
train_dataset = VideoFramesDataset(dataset_path, transform=transform, frames_per_video=8)

# 🔹 تقسيم Train / Val
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_ds, val_ds = torch.utils.data.random_split(train_dataset, [train_size, val_size])

train_loader = DataLoader(train_ds, batch_size=2, shuffle=True, num_workers=2)
val_loader = DataLoader(val_ds, batch_size=2, shuffle=False, num_workers=2)

print(f"✅ Dataset ready! Train: {len(train_ds)}  |  Val: {len(val_ds)}")
```

```python
import torch
from torch import nn, optim
from tqdm import tqdm


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)


epochs = 3
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-5)

model.train()
for epoch in range(epochs):
    running_loss = 0.0
    for frames, labels in tqdm(train_loader, desc=f" Epoch {epoch+1}/{epochs}"):
        frames, labels = frames.to(device), labels.to(device)

        # التأكد من شكل frames
        print("🔹 Original frames shape:", frames.shape)


        if frames.shape[-1] == 3:  
            frames = frames.permute(0, 1, 4, 2, 3)

        print("🔹 Permuted frames shape:", frames.shape)

        outputs = model(frames).logits
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    avg_loss = running_loss / len(train_loader)
    print(f" Epoch {epoch+1} | Average Loss: {avg_loss:.4f}")

print(" Finished Fine-Tuning!")
```
